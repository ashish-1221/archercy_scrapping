```python
import requests

url = "https://en.wikipedia.org/wiki/Archery_at_the_2022_Asian_Games_%E2%80%93_Women%27s_team_recurve"

response = requests.get(url)

if response.status_code == 200:
    html_content = response.text
    print(html_content)
else:
    print(f"Failed to fetch page. Status code: {response.status_code}")
```

```python
from urllib.request import Request, urlopen
def fetch_html(url):
    request = Request(
        url,
        headers={
            "User-Agent": "Mozilla/5.0 (compatible; archery-scraper/1.0)",
        },
    )
    with urlopen(request) as response:
        return response.read().decode("utf-8")
```

```python
html = fetch_html("https://en.wikipedia.org/wiki/Archery_at_the_2022_Asian_Games_%E2%80%93_Women%27s_team_recurve")
```
