# Asian Games 2018

```python
df_1.head()
```

```python
df_1 = df_1.drop(columns=['round_name'],axis=1)
df_1.rename(columns={'round':'round_name'},inplace=True)
```

```python
# Clean the event_name
df_1['event'].unique()
```

```python
event_standardization_dict = {
    "Men's individual recurve": "RECURVE MEN INDIVIDUAL",
    "Men's team recurve": "RECURVE MEN TEAM",
    "Men's team compound": "COMPOUND MEN TEAM",
    "Women's individual recurve": "RECURVE WOMEN INDIVIDUAL",
    "Women's team recurve": "RECURVE WOMEN TEAM",
    "Women's team compound": "COMPOUND WOMEN TEAM",
    'Mixed team recurve': 'RECURVE MIXED TEAM',
    'Mixed team compound': 'COMPOUND MIXED TEAM',
}
df_1['event'] = df_1['event'].map(event_standardization_dict)
```

```python
## Standardizing the round names
round_standardization_dict = {
    'Quarterfinals': 'QUARTERFINAL',
    'Semifinals': 'SEMI-FINAL',
    'Gold medal match': 'GOLD MEDAL MATCH',
    'Bronze medal match': 'BRONZE MEDAL MATCH',
    '1/32 eliminations': '1/32 ELIMINATION',
    '1/16 eliminations': '1/16 ELIMINATION',
    '1/8 eliminations': '1/8 ELIMINATION'
}
df_1['round_name'] = df_1['round_name'].map(round_standardization_dict).fillna(df_1['round_name'])
```

```python
df_1["player_a_countries"] = df_1["player_a"].str.findall(r"\(([^)]*)\)")
df_1["player_a_countries"] =  df_1["player_a_countries"].apply(lambda x: list(set(x)) if isinstance(x, list) else x)


df_1["player_b_countries"] = df_1["player_b"].str.findall(r"\(([^)]*)\)")
df_1["player_b_countries"] =  df_1["player_b_countries"].apply(lambda x: list(set(x)) if isinstance(x, list) else x)

df_1["winner_countries"] = df_1["winner"].str.findall(r"\(([^)]*)\)")
df_1["winner_countries"] =  df_1["winner_countries"].apply(lambda x: list(set(x)) if isinstance(x, list) else x)
```

```python
df_1["player_b"] = df_1["player_b"].str.replace(r"\s*\([^)]*\)", "", regex=True)
df_1["player_b"] = (
    df_1["player_b"]
    .str.replace(r"\s*\([^)]*\)", "", regex=True)
    .str.replace(r"\s*/\s*", " / ", regex=True)  # normalize separators
    .str.strip()
)

df_1["player_a"] = df_1["player_a"].str.replace(r"\s*\([^)]*\)", "", regex=True)
df_1["player_a"] = (
    df_1["player_a"]
    .str.replace(r"\s*\([^)]*\)", "", regex=True)
    .str.replace(r"\s*/\s*", " / ", regex=True)  # normalize separators
    .str.strip()
)


df_1["winner"] = df_1["winner"].str.replace(r"\s*\([^)]*\)", "", regex=True)
df_1["winner"] = (
    df_1["winner"]
    .str.replace(r"\s*\([^)]*\)", "", regex=True)
    .str.replace(r"\s*/\s*", " / ", regex=True)  # normalize separators
    .str.strip()
)
```

```python
## Explode the player_a and player_b columns to have one player per row
df_1 = df_1.assign(
    player_a=df_1["player_a"].str.split(" / ")
).explode("player_a").reset_index(drop=True)

df_1 = df_1.assign(
    player_b=df_1["player_b"].str.split(" / ")
).explode("player_b").reset_index(drop=True)

df_1 = df_1.assign(
    winner=df_1["winner"].str.split(" / ")
).explode("winner").reset_index(drop=True)
```

```python
# df_1 = df_1.explode("countries").reset_index(drop=True)
df_1 = df_1.explode("player_b_countries").reset_index(drop=True)
df_1 = df_1.explode("winner_countries").reset_index(drop=True)
df_1 = df_1.explode("player_a_countries").reset_index(drop=True)
```

```python
df_1['winner_country'] = df_1['winner_countries'].map(COUNTRY__CODES).fillna(df_1['winner_countries'])
df_1['player_b_country'] = df_1['player_b_countries'].map(COUNTRY__CODES).fillna(df_1['player_b_countries'])
df_1['player_a_country'] = df_1['player_a_countries'].map(COUNTRY__CODES).fillna(df_1['player_a_countries'])
```

```python
df_1
```

```python
df_1[['championship_name', 'event', 'round_name', 'player_a','player_a_country','score_a', 'player_b','player_b_country','score_b', 'winner', 'winner_country']].to_csv('asian_games_2018_cleaned.csv', index=False)
```
